```python
# iterate through all 500 customers, stored in a list named `customers`.
# if the customer dict has a 'corporate' key, print the value of the corporate key
# also create a new key-value for the customer called 'account_manager'

# else if customer do not have a corporate key, just print the customer 
# existing bank
for customer in customers:
    # if this key exist in that object
    if 'corporate' in customer:
        print(customer['corporate'])
        # create a new key-value called 'account_manager'
        customer['account_manager'] = 'Samuel'
    # if the customer dict does not have a corporate key
    else: 
        # just assign the customer's bank itself as account manager
        customer['account_manager'] = customer['bank']
```

```python
# for more complicated control flow (a demonstration):
for customer in customers:
    # if this key exist in that object
    if 'corporate' in customer:
        print(customer['corporate'])
        # create a new key-value called 'account_manager'
        customer['account_manager'] = 'Samuel'
    # else, if the customer dict has the bank key
    elif 'bank' in customer:
        # just assign the customer's bank itself as account manager
        customer['account_manager'] = customer['bank']
    # else, if the customer dict has a branch manager
    elif 'branch_mng' in customer:
        customer['account_manager'] = customer['branch_mng']
    # else
    else: 
        # just assign the customer's bank itself as account manager
        customer['account_manager'] = False
```
